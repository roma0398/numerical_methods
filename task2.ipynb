# ИНТЕРПОЛИРОВАНИЕ ФУНКЦИЙ

```python
import numpy as np
import math
import matplotlib.pyplot as plt
```

### задаем функцию $f(x) =x - \sin(x) - 0.25$

```python
def f(x):
    return (x - math.sin(x) - 0.25)
```

### функция вычисляющая полином Лагранжа $L_n(x)=\sum^n_{i=0}\frac{\omega_{n+1}(x)}{(x-x_i)\omega'_{n+1}(x_i)}f(x_i)$

```python
def lagrange(x):
    def l(k):
        w = 1
        for i in range(n):
            if i != k:
                w *= (x - xx[i])
        w_ = 1
        for i in range(n):
            if i != k:
                w_ *= (xx[k] - xx[i])
        return w / w_

    res = 0
    for i in range(n):
        res += yy[i] * l(i)
    return res
```

### Берем равностоящие узлы на промежутке $[-5, 5]$

```python
a = -5
b = 5
n = 10
xx = np.linspace(a, b, n)
yy = np.array(list(map(f, xx)))
xch = np.zeros(n)
```

### Вычисляем узлы по формуле $x_i = \frac{1}{2}\left[(b-a)\cos\frac{(2i+1)}{2(n+1)}\pi+(b+a)\right], i=\overline{1,n}$

```python
for i in range(n):
    xch[i] = 0.5 * ((b - a) * math.cos(
        (2 * i + 1) / (2 * n) * math.pi) + (b + a))
```

### Методическая погрешность

```python
xg = np.arange(a, b + 0.1, 0.1)
maxr = 0
for cur in xch:
    maxr = max(maxr, abs(f(cur) - lagrange(cur)))
print(maxr)
```

### Рисуем график исходного полинома поверх полученного полинома Лагранжа
#### слева график построенный по равноотстоящим узлам, справа по определенных по формуле

```python
plt.figure(figsize=(40, 20))
ych = np.array(list(map(f, xch)))
plt.subplot (1, 2, 2)
plt.plot(xch, ych, linewidth=12, color='k')
ych = np.array(list(map(lagrange, xch)))
plt.grid(ls=':')
plt.plot(xch, ych, linewidth=5, color='y')
yg = np.array(list(map(f, xg)))
plt.subplot (1, 2, 1)
plt.plot(xg, yg, linewidth=12, color='k')
yg = np.array(list(map(lagrange, xg)))
plt.grid(ls=':')
plt.plot(xch, ych, linewidth=5, color='y')
plt.show()
```
